```python
# Plot high temperature

dates = df['DATE']
highs = df['TMAX']

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()

# Plot highs in red
ax.plot(dates, highs, color='red')

# Add titles
ax.set_title("Daily High Temperatures, 2021", fontsize=24)
ax.set_xlabel('', fontsize=16)
plt.xticks(range(0, len(dates), 30)) # Plot every 30th date
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=8)
ax.tick_params(labelsize=16)

plt.show()
```

```python
# Plot lows

lows = df['TMIN']

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()

# Plot highs in red
ax.plot(dates, highs, color='red')

# Plot lows in blue
ax.plot(dates, lows, color='blue')

# Add titles
ax.set_title("Daily High and Low Temperatures, 2021", fontsize=16)
ax.set_xlabel('', fontsize=16)
plt.xticks(range(0, len(dates), 30)) # Plot every 30th date
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=8)
ax.tick_params(labelsize=16)

plt.show()
```

```python
# Shade in area between high and low using fill_between()

plt.style.use('default')

fig, ax = plt.subplots()

# Plot highs in red
ax.plot(dates, highs, color='red', alpha=0.5)

# Plot highs in blue
ax.plot(dates, lows, color='blue', alpha=0.5)

# Add fill
ax.fill_between(dates, highs, lows, facecolor='blue',alpha=0.1)

# Add titles
ax.set_title("Daily High and Low Temperatures, 2021", fontsize=16)
ax.set_xlabel('', fontsize=16)
plt.xticks(range(0, len(dates), 30)) # Plot every 30th date
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=8)
ax.tick_params(labelsize=16)

plt.savefig('high_low.png')

plt.show()
```

## **Plotting earthquake data (GeoJSON)**

```python
# Read in file and write it out in a more readable format

from pathlib import Path
import json # used for structured data

path=Path('eq_data_1_day_m1.geojson')
contents = path.read_text()

all_eq_data = json.loads(contents)

path = Path('readable_eq_data.geojson')

readable_contents=json.dumps(all_eq_data, indent=4)

path.write_text(readable_contents)
```

```python
# prompt: print first 300 elements of readable_contents

# 5 minute exercise
# Explore the JSON immediately above
# print(readable_contents[:300])

# What data does it have? Change 300 to 1000 to show more data
print(readable_contents[:2000])
```

```python
# Data
# Features - data about an earthquake
## mag - magnitude of earthquake (1.6)
## title - magnitude and location of event
# geometry - Where earthquake occurred
## coordinates -  longitude, latitude

# List of all earthquakes in the dataset
all_eq_dicts = all_eq_data['features'] # features of all earthquakes
print(len(all_eq_dicts))               # 160 earthquakes
```

### Extract Hierarchy Objects

```python
# Magnitude of each earthquake

mags = []

for eq_dict in all_eq_dicts:
  mag = eq_dict['properties']['mag'] # The amazing power of JSON and Python!
  mags.append(mag)

print(mags[:10])
```

### Indentify List <u>MAX</u> and <u>MIN</u>

```python
# Print Mags
# print(mags)

# Find the smallest magnitude
# min(mags)

# Find the largest magnitude
max(mags)
```

```python
# Location of each earthquake
mags, lons, lats = [], [], []

for eq_dict in all_eq_dicts:
  mag = eq_dict['properties']['mag'] # The amazing power of JSON and Python!
  lon = eq_dict['geometry']['coordinates'][0]       # Lon-0, lat-1, elevation-2
  lat = eq_dict['geometry']['coordinates'][1]

  mags.append(mag)
  lons.append(lon)
  lats.append(lat)

print(mags[:10])
print(lons[:10])
print(lats[:10])
```

```python
import plotly.express as px # A new library - interactive visualizations

title = 'Global Earthquakes'

fig = px.scatter_geo(lat=lats, lon=lons, title=title)

fig.show()
```

```python
# Represent magnitudes

import plotly.express as px # A new library - interactive visualizations

title = 'Global Earthquakes'

fig = px.scatter_geo(lat=lats, lon=lons, size=mags, title=title)

fig.show()
```

```python
# Use color to highlight big earthquakes

fig = px.scatter_geo(lat=lats, lon=lons, size=mags, title=title,
                     color=mags,                                 # Data for color scale
                     color_continuous_scale='bluered',           # Color scale
                     labels={'color':'Magnitude'},               # Set label to Magnitude
                     projection='natural earth')                 # Change map

fig.show()
```

```python
# Add titles to hover text
mags, lons, lats, eq_titles = [], [], [], []

for eq_dict in all_eq_dicts:
  mag = eq_dict['properties']['mag']
  lon = eq_dict['geometry']['coordinates'][0]
  lat = eq_dict['geometry']['coordinates'][1]
  eq_title = eq_dict['properties']['title']    # Grab title

  mags.append(mag)
  lons.append(lon)
  lats.append(lat)
  eq_titles.append(eq_title)

fig = px.scatter_geo(lat=lats, lon=lons, size=mags, title=title,
                     color=mags,
                     color_continuous_scale='temps',
                     labels={'color':'Magnitude'},
                     projection='natural earth',
                     hover_name=eq_titles)                    # Add titles

fig.show()

```
