```python
# Plot histograms for multiple variables

games.hist(figsize = (10, 6), bins=30, ec='black');

plt.tight_layout()

```

```python
# Exclude largest values because they are distorting the histogram

filter_cutoff = 0.95

filter_price = games['Price'] <=games['Price'].quantile(filter_cutoff)

filter_user_rating_count = games['User Rating Count'] <= games['User Rating Count'].quantile(filter_cutoff)

filter_size = games['Size'] <= games['Size'].quantile(filter_cutoff)

filter_exclude_top_percent = filter_price & filter_user_rating_count & filter_size

games[filter_exclude_top_percent].hist(figsize=(10,6), bins=30, ec='black')

plt.tight_layout()
```

```python
# Categorize the user ratings

ratings_mapping = {1.5: '1_poor', 2:'1_poor', 2.5: '1_poor', 3: '1_poor', 3.5: '2_fair', 4: '2_fair', 4.5: '3_good', 5: '4_excellent'}

games['cat_rating'] = games['Average User Rating'].map(ratings_mapping)

```

```python
# Check new column
games.head(20)
```

```python
# Create a boxplot

sns.boxplot(x='cat_rating', y='Size', data=games, order=['1_poor','2_fair','3_good','4_excellent'])
```

```python
# Limit size to 600MB
sns.boxplot(x='cat_rating', y='Size', data=games[games['Size']<=600], order=['1_poor','2_fair','3_good','4_excellent'])
```

## Extended example

Are highly rated strategy games monolingual or multilingual? Paid or free?

```python
import pandas as pd
```

```python
games = pd.read_csv('appstore_games.csv')
```

```python
games.head()
```

```python
# Clean up column names
# Lower case, add underscores

original_columns_dict = {x: x.lower().replace(' ','_') for x in games.columns}

games.rename(columns = original_columns_dict, inplace=True)
```

```python
games.head()
```

```python
# Set the index to the 'id' column

games.set_index(keys = 'id', inplace=True)
```

```python
games.head()
```

```python
# Eliminate rows where average_user_rating is null

games=games.loc[games['average_user_rating'].notnull()]
```

```python
# Keep games only when user_rating_count >= 30

games = games.loc[games['user_rating_count'] >= 30]
```

```python
games.shape
```

```python
# Impute missing values with string EN

games['languages'] = games['languages'].fillna('EN')
```

```python
# Create new column that indicates whether game is free or paid

games['free_game'] = (games['price'] == 0).astype(int).map({0:'paid', 1:'free'})
```

```python
games.head()
```

```python
# Create new column multilingual
number_of_languages = games['languages'].apply(lambda x: x.count(','))

games['multilingual'] = number_of_languages == 0

games['multilingual'] = games['multilingual'].astype(int).map({0: 'multilingual', 1:'monolingual'})
```

```python
number_of_languages
```

```python
games.head()
```

```python
# One column that combines paid/free and multilingual/monolingual

games['price_language'] = games['free_game'] + '-' + games['multilingual']
```

```python
games.head()
```

```python
# Create table that compares free/paid multilingual/monolingual

games.groupby('price_language')['average_user_rating'].describe()
```
