# Case study: Generating 'random' stock prices

This case study shows how to randomly generate stock charts.

You will learn how to use classes and methods. You will also reinforce what you have learned about drawing charts.

```python
# Our first class!

from random import choice             # Random is a library

class RandomWalk:
  """A class to generate stock market charts."""

  def __init__(self, num_points = 500):
    """Initialize attributes of a stock market chart"""
    self.num_points = num_points

    # Start chart at (0,0)

    self.x_values = [0]
    self.y_values = [100]


  def fill_walk(self):
    """Calculate all the points in the stock market chart"""

    # Keep generating points until reach num_points

    while len(self.x_values) < self.num_points:

      # Set x direction and distance
      x_direction = 1
      x_distance = 1
      x_step = x_direction * x_distance

      # Set y direction and distance
      y_direction = choice([1, 0, -1])
      y_distance = choice([1,2,3,4,5])
      y_step = y_direction * y_distance

      # Calculate new position
      x = self.x_values[-1] + x_step
      y = self.y_values[-1] + y_step

      self.x_values.append(x)
      self.y_values.append(y)
```

```python
# Plot the stock chart!

import matplotlib.pyplot as plt

rw = RandomWalk()   # Create stock market chart
rw.fill_walk()      # Calculate stock prices

plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values)
plt.show()
```

### Improving the visualization of the stock chart

In this section, you learn various ways to modify the stock chart to help business users draw conclusions.

```python
# Apply a colormap

plt.style.use('classic')
fig, ax = plt.subplots()

ax.scatter(rw.x_values, rw.y_values, c=rw.x_values, cmap=plt.cm.Blues, edgecolors='none', s=15)
    # edgecolors - remove colors at edge of point
```

```python
# Highlight start and end points

plt.style.use('classic')
fig, ax = plt.subplots()

point_numbers = range(rw.num_points)

ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15)

# Start point
ax.scatter(rw.x_values[0], rw.y_values[0], c="green", edgecolors='none', s=100)

# End point
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100)

plt.show()
```
