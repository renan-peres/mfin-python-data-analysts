```python
# Goal: Produce a single dataset that lists all student names and their advisors and grades
#
# Challenges
# 1. Students are split it into two datasets (df_full_time and df_part_time)
# 2. Grades are in a separate dataset (df_grades)
# 3. Some students in df_grades may not be in the two student files
```

### Combine two student datasets into one

```python
# Combine student datasets into one

df_students = pd.concat([df_full_time, df_part_time])
df_students
```

```python
# Reset the index!

df_students.reset_index(inplace=True, drop=True)
df_students
```

```python
# Track full-time and part-time status

df_full_time['student_type'] = 'full_time' # create new column
df_part_time['student_type'] = 'part_time' # create new column
pd.concat([df_full_time, df_part_time]).reset_index(drop=True) # drop=True drops the original index
```

### Combine students and grades into one dataset


```python
# Key idea: Combine by matching student ids
# The student id is called a "key"
# The key concept of join!

pd.merge(df_grades,
         df_students,
         left_on='student_id',
         right_on='student_id',
         how='inner')           # include only rows which match
```

```python
# 5 minute exercise
# Replace how='inner' with how='left' and 'how'='right'
# Explain the results
```

```python
pd.merge(df_grades,
         df_students,
         left_on='student_id',
         right_on='student_id',
         how='left')
```

```python
pd.merge(df_grades,
         df_students,
         left_on='student_id',
         right_on='student_id',
         how='outer')
```

```python
# Working with joins
# Produce a list of the names of students who have taken Python with their grades
# df_grades is left table, df_students is the right table

# What is the appropriate join?
```

```python
# Do a left join
# If a student took the course (i.e., is in df_grades), but is not in df_students that is probably an error!

df_result = pd.merge(df_grades,
                     df_students[['student_id','first_name','last_name']],
                     left_on='student_id',
                     right_on='student_id',
                     how='left')

if df_result.first_name.isnull().sum() > 0:
  print('Warning df _students is missing some students')

print(df_result)
df_result = df_result[df_result.first_name.notnull()]
df_result
```

## Constructing the restaurant order set

```python
df_orders.head()
```

```python
# It is hard to read the orders because only the item_id is listed instead of the item_name

df_items
```

```python
# Similarly, only restaurant_id is listed instead of restaurant_name

df_restaurants.head()
```

```python
# Add restaurant name

df_res = (pd.merge(df_orders,
                   df_restaurants[['RESTAURANT_ID','NAME']],
                   left_on='RESTAURANT_ID',
                   right_on='RESTAURANT_ID',
                   how='left')
            .rename(columns={'NAME':'RESTAURANT_NAME'}))

df_res.head() # look at rightmost column


```

```python
# Add name of main

(pd.merge(df_res,
          df_items[['ITEM_ID', 'ITEM_NAME']],
          left_on='MAIN',
          right_on='ITEM_ID',
          how='left')
.rename(columns={'ITEM_NAME':'MAIN_NAME'})).head()
```

```python
# Notice that item_id is also included. That is because every column in the second table is included.
# Let's drop this column at the end

df_res2 = (pd.merge(df_res,
          df_items[['ITEM_ID', 'ITEM_NAME']],
          left_on='MAIN',
          right_on='ITEM_ID',
          how='left')
.rename(columns={'ITEM_NAME':'MAIN_NAME'})
.drop(columns='ITEM_ID'))

df_res2
```

```python
df_res2.MAIN_NAME.value_counts()
```

```python
# What are the most popular bowls?

df_res2.MAIN_NAME.value_counts().plot(kind='bar')
```

```python
# 5 minute exercise
# Repeat the above with BASE
# Change the left_on argument
# Change the name of the column to BASE_NAME
```

```python
df_res3 = (pd.merge(df_res2,
          df_items[['ITEM_ID', 'ITEM_NAME']],
          left_on='BASE',
          right_on='ITEM_ID',
          how='left')
.rename(columns={'ITEM_NAME':'BASE_NAME'})
.drop(columns='ITEM_ID'))

df_res3
```

### Grouping by dates and times using **resample**

```python
# Index must be date/time to use resample

( df_orders.set_index('DATETIME')     # Set index
         .resample('D')               # Resample by day
         .DRINKS                      # Select attribute
         .mean()                      # Calculate mean of DRINKS by day
         .reset_index()
         .head() )
```

```python
# Inputs to resample
# T - minutes
# H - hours
# D - Days
# W - Weeks
# ME - months
# Y - years
# B - Business day

```

```python
# Total number of orders per day at Columbia restaurant

( df_orders
  [df_orders.RESTAURANT_NAME == 'Columbia']
  .set_index('DATETIME')
  .resample('D')
  .size()
  .plot() )
```

```python
# Combining groupby and resample
# Total number of sales per day per store

sales_per_day_per_store = ( df_orders.set_index(df_orders.DATETIME)
                              .groupby('RESTAURANT_NAME')
                              .resample('D')
                              .size()
                              .reset_index()
                            )

```

```python
sales_per_day_per_store.head()
```

```python
# Number of drinks at each restaurant by month

( df_orders.set_index(df_orders.DATETIME)
  .groupby('RESTAURANT_NAME')
  .resample('ME')
  .DRINKS
  .sum()
  .reset_index()
  .head() )
```
